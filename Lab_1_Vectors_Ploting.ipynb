```python
labels = 'Frogs', 'Dogs', 'Cats', 'Birds'# Label for Pie Chart
sizes = [15, 30, 40, 15]#Size for each Piece
explode = (0, 0.1, 0, 0)#Only "explode" the 2nd slice (i.e. 'Dogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)#"autopct prints out the Perct - shadow creats the shadow - startnagle rotates the chart"
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()#Print Chart
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)#Array of Numbers
s = np.sin(2*np.pi*t)#Eq for s(y)

plt.plot(t, s)#Plot
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)#Plot Title
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)#Line Txt for points 1 to -0.6
plt.text(0.5, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize=15)#^
plt.xlabel('time (s)')#X Axis Label
plt.ylabel('volts (mV)')#Y Axis Label
plt.show()#Print Chart
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.colors import Normalize
from matplotlib.markers import MarkerStyle
from matplotlib.text import TextPath
from matplotlib.transforms import Affine2D

SUCCESS_SYMBOLS = [
    TextPath((0, 0), ""),
    TextPath((0, 0), ""),
    TextPath((0, 0), ""),
]

N = 25
skills = np.random.uniform(5, 80, size=N) * 0.1 + 5
takeoff_angles = np.random.normal(0, 90, N)
thrusts = np.random.uniform(size=N)
successful = np.random.randint(0, 3, size=N)
positions = np.random.normal(size=(N, 2)) * 5
data = zip(skills, takeoff_angles, thrusts, successful, positions)

cmap = plt.colormaps["jet"]
fig, ax = plt.subplots()
fig.suptitle("Throwing success", size=14)
for skill, takeoff, thrust, mood, pos in data:
    t = Affine2D().scale(skill).rotate_deg(takeoff)
    m = MarkerStyle(SUCCESS_SYMBOLS[mood], transform=t)
    ax.plot(pos[0], pos[1], marker=m, color=cmap(thrust))


plt.show()
```
